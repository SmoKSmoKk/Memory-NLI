### String operations: remove useless characters

```python
# before cleaning
print(df["title"])
```

```python
# requires additional edditing by hand
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","-"]

# %nl2pandas replace characters
for char in spec_chars:
    df["title"] = df["title"].str.replace(char, ' ')
```

```python
# %nl2pandas split 'title'
df["title"] = df["title"].str.split() 
```

```python
# %nl2pandas join column 'title' on ' '
df["title"] = df["title"].str.join(sep=' ')
```

```python
# after cleaning
print(df["title"])
```

### Extract relevant content

```python
# changed regex from original expression in tutorial
# %nl2pandas extract '(\$\.d*\.?\d*)' from 'title'
df['dollar_prices'] = df["title"].str.extract(pat='(\\$[1-9][0-9]{0,4}(?:\\.\\d+)?|100000(?:\\.0+)?)(?!\\S)')
```

```python
# %nl2pandas display unique values on column 'dollar_prices'
df["dollar_prices"].unique()
```

```python
# %nl2pandas find null values
df.isnull()
```

### Check for missing values

```python
# %nl2pandas remove missing values from 'dollar_prices'
df = df.dropna(subset='dollar_prices')
```

### Change datatype

```python
# %nl2pandas replace "$" with " " on "dollar_prices"
df["dollar_prices"] = df["dollar_prices"].str.replace(pat='$', repl=' ')
```

```python
# dollor prices is type object
df.dtypes
```

```python
# %nl2pandas change the type of 'df["dollar_prices"]' to numeric
df['dollar_prices'] = pd.to_numeric(arg=df['dollar_prices'])
```
